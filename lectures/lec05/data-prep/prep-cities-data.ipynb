The data comes from here:   
    - https://www.macrotrends.net/cities/23129/san-diego/population  
    - https://www.macrotrends.net/cities/23131/san-jose/population  
You may just need to add one new row each new year.

```python
import pandas as pd
```

```python
sd = pd.read_csv('San Diego-population-2022-01-20.csv', header=9)
sj = pd.read_csv('San Jose-population-2022-01-20.csv', header=9)
```

```python
sd['date'] = sd['date'].str[:4]
sj['date'] = sj['date'].str[:4]
```

```python
sd
```

```python
sd['Pop SD'] = sd[' Population']
sd['Pop SJ'] = sj[' Population']
sd['Growth SD'] = sd[' Annual Change']
sd['Growth SJ'] = sj[' Annual Change']
```

```python
final = sd[['date', 'Pop SD', 'Growth SD', 'Pop SJ', 'Growth SJ']].iloc[20:-13]
final
```

```python
final.to_csv('../data/sd-sj-2022.csv', index=False)
```
